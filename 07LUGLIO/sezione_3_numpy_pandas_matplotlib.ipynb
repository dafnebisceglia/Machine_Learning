### Esercizio 1.5 — Addizione vettoriale: confronto tra due serie di dati

Quando si analizzano dati nel tempo, è comune **confrontare due serie di misurazioni** sommando o sottraendo i vettori corrispondenti elemento per elemento — senza scrivere nessun ciclo `for`.

Date le vendite mensili (in unità) di un negozio nel 2022 e nel 2023, calcola:

1. La **variazione assoluta mensile**: quante unità sono cambiate ogni mese (in valore assoluto, senza segno) → `variazione`
2. La **media mensile** tra i due anni per ogni mese → `media_mensile`
3. La **variazione assoluta totale** sull'anno: la somma di tutte le variazioni assolute mensili → `crescita_totale`

**Suggerimento:** in NumPy, `+`, `-` e `/` tra due array (o tra un array e uno scalare) operano **elemento per elemento**. Per il valore assoluto usa `np.abs()`.


````python
vendite_2022 = np.array([120, 135, 150, 110, 145, 180, 200, 195, 160, 130, 115, 175])
vendite_2023 = np.array([130, 128, 165, 125, 152, 190, 215, 205, 148, 142, 125, 188])

# Variazione assoluta mese per mese: quante unità sono cambiate (senza segno)
variazione = None  # SCRIVI IL TUO CODICE QUI

# Media mensile tra i due anni per ogni mese
media_mensile = None  # SCRIVI IL TUO CODICE QUI

# Crescita totale: somma di tutte le variazioni mensili
crescita_totale = None  # SCRIVI IL TUO CODICE QUI
````

````python
# ✅ Test Esercizio 1.5
attesa_variazione = np.abs(vendite_2023 - vendite_2022)
attesa_media      = (vendite_2022 + vendite_2023) / 2
attesa_crescita   = int(attesa_variazione.sum())  # 136

ok_var  = isinstance(variazione,    np.ndarray) and np.allclose(variazione,    attesa_variazione)
ok_med  = isinstance(media_mensile, np.ndarray) and np.allclose(media_mensile, attesa_media)
ok_cres = crescita_totale is not None and np.isclose(float(crescita_totale), attesa_crescita)

casi = [
    ("variazione",     ok_var,  variazione,    attesa_variazione),
    ("media_mensile",  ok_med,  media_mensile, attesa_media),
    ("crescita_totale", ok_cres, crescita_totale, attesa_crescita),
]
for nome, ok, trovato, atteso_val in casi:
    if ok:
        print(f"✅ {nome} = {trovato}")
    else:
        print(f"❌ {nome} = {trovato}  (atteso: {atteso_val})")
````

````python
quantita = np.array([2, 3, 6, 1])
prezzi   = np.array([1.50, 1.20, 0.40, 0.90])

# Calcola il totale della spesa con un prodotto scalare
costo_totale = None  # SCRIVI IL TUO CODICE QUI

print("Totale spesa: €", costo_totale)
````

````python
# ✅ Test Esercizio 1.5
# 2×1.50 + 3×1.20 + 6×0.40 + 1×0.90 = 3.00 + 3.60 + 2.40 + 0.90 = 9.90
atteso = 9.90
if round(float(costo_totale), 2) == atteso:
    print(f"✅ Corretto! Totale spesa: € {costo_totale:.2f}")
else:
    print(f"❌ Totale: € {costo_totale:.2f}  (atteso: € {atteso:.2f})")
````

## Parte 5 — Riepilogo Finale: Dove Vive la Classe?

Hai imparato NumPy, Pandas e Matplotlib separatamente. Ora le usi **tutte e tre insieme** in un unico workflow realistico — esattamente come si fa in un progetto di analisi dati.

**La storia:** un professore vuole capire da dove vengono i suoi studenti. Ha raccolto le **coordinate geografiche** (latitudine e longitudine) delle loro città di origine e vuole trovare il **baricentro geografico** della classe: il punto centrale attorno a cui si distribuisce il gruppo.

> **Coordinate geografiche in breve:**
> - **Latitudine** (°N): quanto sei a nord. L'Italia va da ~37° (Sicilia) a ~47° (Alto Adige).
> - **Longitudine** (°E): quanto sei a est. L'Italia va da ~7° (Liguria) a ~18° (Puglia).
> - **Distanza approssimata:** ogni grado corrisponde a circa **111 km** in entrambe le direzioni.

La distanza di un punto dal baricentro si calcola con il **teorema di Pitagora** applicato alla griglia lat/lon:

$$d \approx \sqrt{(\Delta\text{lat})^2 + (\Delta\text{lon})^2} \times 111 \quad \text{km}$$
