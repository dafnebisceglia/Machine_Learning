### Esercizio 1.2 — Conta le parole di una frase

Scrivi una funzione `conta_parole(frase)` che riceve una stringa e restituisce il **numero di parole** che contiene.

Esempio: `"Python è fantastico"` → `3`

**Suggerimenti:**
- Prima trasforma la frase in una collezione di parole.
- Poi conta quanti elementi contiene quella collezione.
- Prova anche con una frase che contiene più spazi tra una parola e l'altra.

````python
def conta_parole(frase):
    # SCRIVI IL TUO CODICE QUI
    
    lista_parole = frase.split()
    totale = len(lista_parole)
    return totale

conta_parole("vaffa jnfj kwjh kwjdqn fjknfj")
````

### Esercizio 1.3 — Costruisci le iniziali

Scrivi una funzione `iniziali(nome, cognome)` che riceve nome e cognome come stringhe separate e restituisce le iniziali nel formato `"M.R."` (lettera maiuscola seguita da punto).

Esempio: `iniziali("mario", "rossi")` → `"M.R."`

**Suggerimenti:**
- Ti serve solo il primo carattere di ciascuna stringa.
- Le iniziali devono essere maiuscole anche se l'input è minuscolo.
- Costruisci la stringa finale combinando lettere e punti nel formato richiesto.

````python
def iniziali(nome, cognome):
    # SCRIVI IL TUO CODICE QUI    
    fnome = nome[0].capitalize()
    fcognome = cognome[0].capitalize()
    return f"{fnome}.{fcognome}."
    
iniziali("dafne", "bisceglia")
````

## Parte 2 — `int` e `float`: Numeri

- `int` rappresenta numeri interi: indici, contatori, risultati di divisioni intere.
- `float` rappresenta numeri con virgola: misure, prezzi, percentuali.

| Operatore/Funzione | Significato |
|--------------------|-------------|
| `+` `-` `*` `/` | Operazioni aritmetiche di base |
| `**` | Elevamento a potenza |
| `%` | Resto della divisione (modulo) |
| `round(n, cifre)` | Arrotonda `n` a `cifre` decimali |
| `f"{n:.2f}"` | Formatta `n` con 2 cifre decimali |

### Esercizio 2.1 — Resto della divisione

Scrivi una funzione `e_multiplo(n, divisore)` che restituisce `True` se `n` è un multiplo esatto di `divisore`, altrimenti `False`.

Esempio: `e_multiplo(12, 4)` → `True` &nbsp;&nbsp; `e_multiplo(7, 3)` → `False`

**Suggerimenti:**
- Ragiona sul resto della divisione tra `n` e `divisore`.
- Il risultato della funzione deve essere un valore booleano.
- Non serve stampare nulla: la funzione deve restituire il risultato.

````python
def e_multiplo(n, divisore):
    # SCRIVI IL TUO CODICE QUI
    if n % divisore == 0:
        return True
    else:
        return False

e_multiplo(4, 12)
````

Scrivi una funzione `formatta_euro(prezzo)` che riceve un numero `float` e restituisce una stringa con il prezzo formattato con esattamente **due cifre decimali** e il simbolo `€`.

## Parte 3 — `list`: Liste

Una lista è una sequenza **mutabile** e **ordinata** di valori. Gli elementi sono indicizzati a partire da `0`.

| Operazione | Esempio |
|------------|---------|
| Accesso per indice | `lista[0]`, `lista[-1]` (ultimo) |
| Slicing | `lista[1:3]` (dall'indice 1 al 2 incluso) |
| Aggiunta | `lista.append(elemento)` |
| Lunghezza | `len(lista)` |
| Lista ordinata | `sorted(lista)` (non modifica l'originale) |

### Esercizio 3.1 — Aggiungi e ordina

Scrivi una funzione `aggiungi_e_ordina(lista, elemento)` che aggiunge `elemento` alla lista e restituisce una **nuova lista ordinata** (senza modificare quella originale).

Esempio: `aggiungi_e_ordina(["Ron", "Harry"], "Hermione")` → `["Harry", "Hermione", "Ron"]`

**Suggerimenti:**
- Attenzione alla differenza tra modificare una lista esistente e creare una nuova lista.
- Costruisci prima la sequenza che contiene anche il nuovo elemento, poi ordina il risultato.
- La lista passata come input deve rimanere identica dopo la chiamata della funzione.

````python
def aggiungi_e_ordina(lista, elemento):
    nuova_lista = lista.copy()
    nuova_lista.append(elemento)
    nuova_lista.sort()

    return nuova_lista

risultato = aggiungi_e_ordina(["vc", "sg", "dsg"], "inter")
print(risultato)
````

````python
# ✅ Test Esercizio 3.1 — dovresti vedere due righe con "✅ Corretto!"
casi = [
    (["Ron", "Harry"],   "Hermione", ["Harry", "Hermione", "Ron"]),
    ([3, 1, 4],          2,          [1, 2, 3, 4]),
]
for lista, elemento, atteso in casi:
    risultato = aggiungi_e_ordina(lista[:], elemento)  # lista[:] per non alterare i dati del test
    if risultato == atteso:
        print(f"✅ Corretto!  aggiungi_e_ordina({lista}, {elemento!r}) → {risultato}")
    else:
        print(f"❌ aggiungi_e_ordina({lista}, {elemento!r}) → {risultato}  (atteso: {atteso})")
````

### Esercizio 3.2 — Primo e ultimo elemento

Scrivi una funzione `primo_e_ultimo(lista)` che restituisce una **tupla** con il primo e l'ultimo elemento della lista.

Esempio: `primo_e_ultimo([10, 20, 30, 40])` → `(10, 40)`

**Suggerimenti:**
- Usa gli indici per accedere a elementi specifici della lista.
- Python permette di contare anche a partire dalla fine della lista.
- Il risultato deve essere una tupla di due elementi, non una lista.

````python
def primo_e_ultimo(lista):
    # SCRIVI IL TUO CODICE QUI   
    primo = lista[0]
    secondo = lista[-1]
    tup = (primo, secondo)
    return tup
````

### Esercizio 3.3 — Estrai una sottosequenza (slicing)

Scrivi una funzione `sottosequenza(lista, da, a)` che restituisce gli elementi della lista compresi tra l'indice `da` (incluso) e l'indice `a` (escluso).

Esempio: `sottosequenza([10, 20, 30, 40, 50], 1, 4)` → `[20, 30, 40]`

**Suggerimenti:**
- Questo è un caso d'uso della sintassi di slicing.
- Controlla con attenzione quale estremo è incluso e quale è escluso.
- La funzione deve restituire una nuova sottolista.

````python
def sottosequenza(lista, da, a):
    # SCRIVI IL TUO CODICE QUI
    risultato = lista[da:a]
    return risultato
````

### Esercizio 3.4 — List comprehension: filtra e trasforma

Scrivi una funzione `quadrati_pari(numeri)` che riceve una lista di interi e restituisce una **nuova lista** con i quadrati dei soli numeri **pari**.

Esempio: `quadrati_pari([1, 2, 3, 4, 5, 6])` → `[4, 16, 36]`

**Suggerimenti:**
- Pensa a come descriveresti la regola in una frase: "voglio *questa trasformazione* per ogni elemento *che soddisfa questa condizione*".
- Python offre una sintassi compatta per costruire liste che segue esattamente questa struttura logica, senza scrivere un ciclo esplicito con `append`.
- Se non ti viene subito, scrivi prima la versione con `for` e `if`, poi prova a condensarla in una singola riga.

````python
def quadrati_pari(numeri):
    # SCRIVI IL TUO CODICE QUI
    solo_pari = [n**2 for n in numeri if n % 2 == 0]
    return solo_pari 
````

````python
# ✅ Test Esercizio 3.4 — dovresti vedere tre righe con "✅ Corretto!"
casi = [
    ([1, 2, 3, 4, 5, 6], [4, 16, 36]),
    ([1, 3, 5],           []),
    ([2, 4, 6],           [4, 16, 36]),
]
for numeri, atteso in casi:
    risultato = quadrati_pari(numeri)
    if risultato == atteso:
        print(f"✅ Corretto!  quadrati_pari({numeri}) → {risultato}")
    else:
        print(f"❌ quadrati_pari({numeri}) → {risultato}  (atteso: {atteso})")
````

## Parte 4 — `tuple`: Tuple

Le tuple sono sequenze **ordinate** e **immutabili**: una volta create, i loro elementi non possono essere modificati.
Sono utili per rappresentare coppie o gruppi di valori che devono restare stabili (es. coordinate, risultati multipli di una funzione).

```python
# Creazione
coordinate = (10, 20)

# Unpacking: assegna ogni elemento a una variabile separata

```

### Esercizio 4.1 — Crea e spacchetta una tupla

Scrivi una funzione `distanza_dall_origine(punto)` che riceve una tupla `(x, y)` e restituisce la **distanza dal punto di origine** `(0, 0)` usando la formula:

$$d = \sqrt{x^2 + y^2}$$

Esempio: `distanza_dall_origine((3, 4))` → `5.0`

**Suggerimenti:**
- Prima estrai le due coordinate dalla tupla.
- Poi traduci la formula in operazioni Python.
- Ricorda che il risultato deve essere un numero, non una stringa.

````python

def distanza_dall_origine(punto): 
    # SCRIVI IL TUO CODICE QUI
    x, y = punto
    risultato = ((x**2) + (y**2)) ** (1/2)
    
    return risultato
````

````python
# ✅ Test Esercizio 4.1 — dovresti vedere tre righe con "✅ Corretto!"
casi = [((3, 4), 5.0), ((0, 0), 0.0), ((6, 8), 10.0)]
for punto, atteso in casi:
    risultato = distanza_dall_origine(punto)
    if round(risultato, 5) == atteso:
        print(f"✅ Corretto!  distanza_dall_origine({punto}) → {risultato}")
    else:
        print(f"❌ distanza_dall_origine({punto}) → {risultato}  (atteso: {atteso})")
````

### Esercizio 4.2 — Funzione con valori di ritorno multipli

In Python, una funzione può restituire più valori separandoli con una virgola: Python li impacchetta automaticamente in una tupla.

Scrivi una funzione `minimo_e_massimo(lista)` che riceve una lista di numeri e restituisce **due valori**: il minimo e il massimo.

Esempio: `minimo_e_massimo([3, 1, 7, 2])` → `(1, 7)`

**Suggerimenti:**
- Puoi usare le funzioni built-in per individuare gli estremi di una lista.
- Restituisci i due risultati nello stesso `return`, nell'ordine richiesto: prima minimo, poi massimo.

````python
def minimo_e_massimo(lista):
    # SCRIVI IL TUO CODICE QUI
    minimo = min(lista)
    massimo = max(lista)
    return minimo, massimo
````

````python
# ✅ Test Esercizio 4.2 — dovresti vedere due righe con "✅ Corretto!"
# Nota: puoi fare unpacking del risultato → minimo, massimo = minimo_e_massimo([...])
casi = [([3, 1, 7, 2], (1, 7)), ([10, 10, 10], (10, 10))]
for lista, atteso in casi:
    risultato = minimo_e_massimo(lista)
    if risultato == atteso:
        print(f"✅ Corretto!  minimo_e_massimo({lista}) → {risultato}")
    else:
        print(f"❌ minimo_e_massimo({lista}) → {risultato}  (atteso: {atteso})")
````

## Parte 5 — `dict`: Dizionari

Un dizionario associa **chiavi** a **valori**. È la struttura ideale per rappresentare oggetti con attributi etichettati (record, JSON, configurazioni).

```python
# Creazione
studente = {"nome": "Harry", "voto": 28}

# Accesso
nome = studente["nome"]

# Modifica / aggiunta
studente["voto"] = 30

# Accesso sicuro (evita KeyError se la chiave non esiste)
citta = studente.get("citta")   # restituisce None se non trovata
```

### Esercizio 5.1 — Costruisci un record studente

Scrivi una funzione `crea_studente(nome, corso, voto)` che riceve nome, corso e voto e restituisce un **dizionario** con queste tre chiavi.

Esempio: `crea_studente("Harry", "Gryffindor", 28)` → `{"nome": "Harry", "corso": "Gryffindor", "voto": 28}`

**Suggerimenti:**
- Ogni informazione deve essere accessibile tramite una chiave descrittiva.
- I nomi delle chiavi devono corrispondere a quelli mostrati nell'esempio.
- La funzione deve costruire e restituire il record, non stamparlo.

````python
def crea_studente(nome, corso, voto):
    # SCRIVI IL TUO CODICE QUI
    dizionario = {"nome" : nome, "corso" : corso, "voto" : voto}
    return dizionario
````

````python
# ✅ Test Esercizio 5.1 — dovresti vedere due righe con "✅ Corretto!"
casi = [
    ("Harry",  "Gryffindor", 28, {"nome": "Harry",  "corso": "Gryffindor", "voto": 28}),
    ("Draco",  "Slytherin",  25, {"nome": "Draco",  "corso": "Slytherin",  "voto": 25}),
]
for nome, corso, voto, atteso in casi:
    risultato = crea_studente(nome, corso, voto)
    if risultato == atteso:
        print(f"✅ Corretto!  crea_studente({nome!r}, {corso!r}, {voto}) → {risultato}")
    else:
        print(f"❌ crea_studente({nome!r}, {corso!r}, {voto}) → {risultato}  (atteso: {atteso})")
````

### Esercizio 5.2 — Aggiorna un campo del dizionario

Scrivi una funzione `promuovi_studente(studente, nuovo_voto)` che riceve un dizionario studente e un nuovo voto, **aggiorna** il campo `"voto"` nel dizionario e restituisce il dizionario aggiornato.

Esempio: `promuovi_studente({"nome": "Ron", "voto": 18}, 24)` → `{"nome": "Ron", "voto": 24}`

**Suggerimenti:**
- Individua nel dizionario la chiave che rappresenta il voto.
- Aggiorna quel valore mantenendo invariati gli altri campi.
- Restituisci il dizionario dopo l'aggiornamento.

````python
def promuovi_studente(studente, nuovo_voto):
    # SCRIVI IL TUO CODICE QUI
    studente["voto"] = nuovo_voto
    return studente
````

````python
# ✅ Test Esercizio 5.2 — dovresti vedere due righe con "✅ Corretto!"
casi = [
    ({"nome": "Ron",      "voto": 18}, 24, {"nome": "Ron",      "voto": 24}),
    ({"nome": "Hermione", "voto": 28}, 30, {"nome": "Hermione", "voto": 30}),
]
for studente, nuovo_voto, atteso in casi:
    risultato = promuovi_studente(studente.copy(), nuovo_voto)
    if risultato == atteso:
        print(f"✅ Corretto!  voto aggiornato a {nuovo_voto} → {risultato}")
    else:
        print(f"❌ risultato: {risultato}  (atteso: {atteso})")
````

## Parte 6 — `set`: Insiemi

Un set è una collezione di valori **unici** e non ordinata. È lo strumento ideale per eliminare duplicati o verificare l'appartenenza di un valore.

```python
colori = set()           # set vuoto
colori.add("rosso")      # aggiunge un elemento
colori.add("rosso")      # ignorato: già presente

# Conversione da lista a set (deduplicazione)
unici = set([1, 2, 2, 3, 3])   # {1, 2, 3}
```

### Esercizio 6.1 — Rimuovi i duplicati da una lista

Scrivi una funzione `valori_unici(lista)` che riceve una lista e restituisce un **set** contenente solo i valori unici.

Esempio: `valori_unici(["A", "B", "A", "C", "B"])` → `{"A", "B", "C"}`

**Suggerimenti:**
- Scegli la struttura dati Python che non ammette duplicati.
- Non serve conservare l'ordine originale.
- La funzione deve restituire la collezione di valori unici.

````python
def valori_unici(lista):
    # SCRIVI IL TUO CODICE QUI
    val_unici = set(lista)
    return val_unici
````

````python
# ✅ Test Esercizio 6.1 — dovresti vedere due righe con "✅ Corretto!"
casi = [
    (["A", "B", "A", "C", "B"],    {"A", "B", "C"}),
    ([1, 2, 2, 3, 3, 3],            {1, 2, 3}),
]
for lista, atteso in casi:
    risultato = valori_unici(lista)
    if risultato == atteso:
        print(f"✅ Corretto!  valori_unici({lista}) → {risultato}")
    else:
        print(f"❌ valori_unici({lista}) → {risultato}  (atteso: {atteso})")
````

### Esercizio 6.2 — Quante categorie distinte?

Scrivi una funzione `conta_categorie(lista)` che riceve una lista di valori (potenzialmente con duplicati) e restituisce il **numero di valori distinti** presenti.

Esempio: `conta_categorie(["Gryffindor", "Slytherin", "Gryffindor", "Hufflepuff"])` → `3`

**Suggerimenti:**
- Prima ragiona su come ottenere solo i valori distinti.
- Poi conta quanti valori rimangono.
- Il risultato finale deve essere un numero intero.

````python
def conta_categorie(lista):
    # SCRIVI IL TUO CODICE QUI
    valori_unici = set(lista)
    lunghezza = len(valori_unici)
    return lunghezza
````

````python
# ✅ Test Esercizio 6.2 — dovresti vedere due righe con "✅ Corretto!"
casi = [
    (["Gryffindor", "Slytherin", "Gryffindor", "Hufflepuff"], 3),
    (["rosso", "rosso", "rosso"],                              1),
]
for lista, atteso in casi:
    risultato = conta_categorie(lista)
    if risultato == atteso:
        print(f"✅ Corretto!  conta_categorie({lista}) → {risultato}")
    else:
        print(f"❌ conta_categorie({lista}) → {risultato}  (atteso: {atteso})")
````

### Esercizio 6.3 — Elementi comuni a due collezioni

Scrivi una funzione `elementi_comuni(lista_a, lista_b)` che riceve due liste e restituisce un **set** contenente solo gli elementi presenti in **entrambe**.

Esempio: `elementi_comuni([1, 2, 3, 4], [3, 4, 5, 6])` → `{3, 4}`

**Suggerimenti:**
- Pensa all'operazione della teoria degli insiemi che produce un insieme con solo gli elementi condivisi da entrambi.
- Prima di poter usare le operazioni tra insiemi, converti le liste nella struttura dati appropriata.
- Python supporta le operazioni insiemistiche classiche sia tramite operatori che tramite metodi — sperimenta.

````python
def elementi_comuni(lista_a, lista_b):
    # SCRIVI IL TUO CODICE QUI
    agg1 = set(lista_a)
    agg2 = set(lista_b)
    intersezione_liste = agg1 & agg2
    return intersezione_liste
````

````python
# ✅ Test Esercizio 6.3 — dovresti vedere tre righe con "✅ Corretto!"
casi = [
    ([1, 2, 3, 4],                 [3, 4, 5, 6],    {3, 4}),
    (["Harry", "Ron", "Hermione"], ["Ron", "Draco"], {"Ron"}),
    ([1, 2],                       [3, 4],           set()),
]
for lista_a, lista_b, atteso in casi:
    risultato = elementi_comuni(lista_a, lista_b)
    if risultato == atteso:
        print(f"✅ Corretto!  elementi_comuni({lista_a}, {lista_b}) → {risultato}")
    else:
        print(f"❌ elementi_comuni({lista_a}, {lista_b}) → {risultato}  (atteso: {atteso})")
````

## Parte 7 — `None`: Assenza di valore

`None` è un tipo speciale (`NoneType`) che rappresenta l'**assenza intenzionale** di un valore.
Viene usato per inizializzare variabili senza contenuto e per gestire dati mancanti.

```python
# Le funzioni senza return esplicito restituiscono None
risultato = print("ciao")   # risultato è None

# .get() su un dizionario restituisce None se la chiave non esiste
voto = voti.get("Chimica")  # None se la chiave manca

# Controllo corretto: usa  is None  (non ==)
if voto is None:
    print("Esame non sostenuto")
```

### Esercizio 7.1 — Cerca un voto nel registro

Scrivi una funzione `cerca_voto(registro, materia)` che riceve un dizionario `registro` (materia → voto) e una stringa `materia`, e restituisce:

- il voto (numero) se la materia è presente nel registro
- la stringa `"Non sostenuto"` se la materia **non** è nel registro

Esempio:

```python
registro = {"Matematica": 28, "Fisica": 30}
cerca_voto(registro, "Fisica")   # → 30
cerca_voto(registro, "Chimica")  # → "Non sostenuto"
```

**Suggerimenti:**
- Un dizionario permette di cercare un valore a partire da una chiave.
- Devi gestire anche il caso in cui la chiave richiesta non esiste.
- Usa il valore speciale visto nella sezione su `None` per distinguere un risultato assente da un voto trovato.

````python
def cerca_voto(registro, materia):
    # SCRIVI IL TUO CODICE QUI
    voto = registro.get(materia, "Non sostenuto")
    return voto
````

````python
# ✅ Test Esercizio 7.1 — dovresti vedere quattro righe con "✅ Corretto!"
registro = {"Matematica": 28, "Fisica": 30, "Inglese": 25}
casi = [
    ("Fisica",      30),
    ("Matematica",  28),
    ("Chimica",     "Non sostenuto"),
    ("Storia",      "Non sostenuto"), 
]
for materia, atteso in casi:
    risultato = cerca_voto(registro, materia)
    if risultato == atteso:
        print(f"✅ Corretto!  cerca_voto(registro, {materia!r}) → {risultato!r}")
    else:
        print(f"❌ cerca_voto(registro, {materia!r}) → {risultato!r}  (atteso: {atteso!r})")
````

### Esercizio 7.2 — Filtra i valori mancanti

Scrivi una funzione `filtra_none(lista)` che riceve una lista che può contenere valori `None` e restituisce una **nuova lista** con i soli elementi che non sono `None`, mantenendo l'ordine originale.

Esempio: `filtra_none([10, None, 20, None, 30])` → `[10, 20, 30]`

**Suggerimenti:**
- Ricorda come si controlla correttamente l'identità di `None` (vedi la sezione introduttiva di questa parte).
- Devi costruire una nuova lista: non modificare quella in ingresso.
- Pensa a quale struttura usi ogni volta che vuoi costruire una lista selezionando solo certi elementi.

````python
def filtra_none(lista):
    nuova_lista = []
    for valore in lista:
        if valore is not None:
            nuova_lista.append(valore)
        else:
            continue
    return nuova_lista
````

````python
# ✅ Test Esercizio 7.2 — dovresti vedere tre righe con "✅ Corretto!"
casi = [
    ([10, None, 20, None, 30],        [10, 20, 30]),
    ([None, None],                     []),
    (["alfa", "beta", None, "gamma"],  ["alfa", "beta", "gamma"]),
]
for lista, atteso in casi:
    risultato = filtra_none(lista)
    if risultato == atteso:
        print(f"✅ Corretto!  filtra_none({lista}) → {risultato}")
    else:
        print(f"❌ filtra_none({lista}) → {risultato}  (atteso: {atteso})")
````
