## Parte 1 — Condizionali

I condizionali permettono al programma di eseguire blocchi di codice diversi in base al valore di una condizione booleana.

| Costrutto | Significato |
|-----------|-------------|
| `if condizione:` | Esegue il blocco se la condizione è vera |
| `elif altra_condizione:` | Controlla una seconda condizione (solo se la prima era falsa) |
| `else:` | Esegue il blocco in tutti i casi rimanenti |

**Operatori di confronto:** `>` `<` `>=` `<=` `==` `!=`  
**Operatori logici:** `and`, `or`, `not`

### Esercizio 1.1 — Segno di un numero

Scrivi una funzione `classifica_numero(n)` che riceve un numero intero e restituisce:
- `"Positivo"` se il numero è maggiore di zero
- `"Negativo"` se il numero è minore di zero
- `"Zero"` se il numero è uguale a zero

**Suggerimenti:**
- Usa `if`, `elif` ed `else` per coprire tutti e tre i casi.
- Confronta `n` con `0` usando gli operatori `>`, `<` e `==`.
- Ricorda di usare `return` per restituire la stringa risultante.

````python
def classifica_numero(n):
    # SCRIVI IL TUO CODICE QUI
    if n > 0:
        return "Positivo"
    elif n < 0:
        return "Negativo"
    else:
        return "Zero"
    
````

### Esercizio 1.2 — Classifica il voto universitario

Scrivi una funzione `classifica_voto(voto)` che riceve un voto (numero intero da 0 a 30) e restituisce la corrispondente valutazione testuale:

| Voto | Valutazione |
|------|-------------|
| < 18 | `"Non sufficiente"` |
| 18 – 23 | `"Sufficiente"` |
| 24 – 27 | `"Buono"` |
| 28 – 30 | `"Ottimo"` |

**Suggerimenti:**
- Usa `if` per il primo caso, poi `elif` per ciascuno degli altri.
- Ricorda che non appena una condizione `if` o `elif` è vera, Python salta tutte quelle successive: puoi sfruttarlo per semplificare le condizioni.
- Esempio: se sai già che il voto è `>= 18`, il prossimo controllo può essere `voto <= 23` senza ripetere `>= 18`.

````python
def classifica_voto(voto):
    # SCRIVI IL TUO CODICE QUI
    if voto < 18:
        return "Non sufficiente"
    elif voto >= 18 and voto <= 23:
        return "Sufficiente"
    elif voto >= 24 and voto <= 27:
        return "Buono"
    elif voto >= 28 and voto <= 30:
        return "Ottimo"
````

````python
# ✅ Test Esercizio 1.2 — dovresti vedere quattro righe con "✅ Corretto!"
casi = [(15, "Non sufficiente"), (20, "Sufficiente"), (25, "Buono"), (30, "Ottimo")]
for voto, atteso in casi:
    risultato = classifica_voto(voto)
    if risultato == atteso:
        print(f"✅ Corretto!  classifica_voto({voto}) → '{risultato}'")
    else:
        print(f"❌ classifica_voto({voto}) → '{risultato}'  (atteso: '{atteso}')")
````

### Esercizio 1.3 — Accesso consentito?

Scrivi una funzione `accesso_consentito(eta, ha_account)` che restituisce `True` se l'utente può accedere alla piattaforma, altrimenti `False`.

**Suggerimenti:**

## Parte 2 — Ciclo `for`

Il ciclo `for` si usa quando devi ripetere un'azione su ogni elemento di una sequenza (lista, stringa, range...).

```python
# Itera su ogni elemento di una lista
for elemento in lista:
    # blocco eseguito per ogni elemento

# Itera un numero fissato di volte con range()
for i in range(5):   # i assume i valori 0, 1, 2, 3, 4
    # blocco eseguito 5 volte
```

**Operatore utile:** `%` (resto della divisione) — serve per verificare se un numero è pari: `n % 2 == 0`.

### Esercizio 2.1 — Somma gli elementi di una lista

Scrivi una funzione `somma_lista(numeri)` che riceve una lista di numeri e restituisce la loro somma, **senza usare la funzione built-in `sum()`**.

**Suggerimenti:**
- Ti serve una variabile che conservi il totale mentre attraversi la lista.
- Un ciclo `for` è adatto quando devi visitare tutti gli elementi di una collezione.
- Alla fine della funzione restituisci il totale calcolato.

````python
def somma_lista(numeri):
    # SCRIVI IL TUO CODICE QUI
    somma = 0
    for i in numeri:
        somma += i
    return somma
````

### Esercizio 2.2 — Conta i numeri pari

Scrivi una funzione `conta_pari(numeri)` che riceve una lista di interi e restituisce **quanti** di essi sono numeri pari.

**Suggerimenti:**
- Distingui due lavori: riconoscere se un singolo numero è pari e contare quante volte succede.
- Usa un ciclo per controllare tutti gli elementi della lista.
- Restituisci il conteggio finale, non la lista dei numeri pari.

````python
def conta_pari(numeri):
    # SCRIVI IL TUO CODICE QUI
    contatore = 0
    for valore in numeri:
        if valore % 2 == 0:
            contatore += 1
    return contatore
````

````python
# ✅ Test Esercizio 2.2 — dovresti vedere due righe con "✅ Corretto!"
casi = [([1, 2, 3, 4, 5, 6], 3), ([7, 11, 13], 0)]
for lista, atteso in casi:
    risultato = conta_pari(lista)
    if risultato == atteso:
        print(f"✅ Corretto!  conta_pari({lista}) → {risultato}")
    else:
        print(f"❌ conta_pari({lista}) → {risultato}  (atteso: {atteso})")
````

### Esercizio 2.3 — Trova il valore massimo

Scrivi una funzione `trova_massimo(numeri)` che riceve una lista di numeri e restituisce il valore più grande, **senza usare la funzione built-in `max()`**.

**Suggerimenti:**
- Mantieni in memoria il miglior candidato trovato finora.
- Confronta ogni nuovo elemento con quel candidato mentre attraversi la lista.
- Puoi assumere che la lista non sia vuota.

````python
def trova_massimo(numeri):
    # SCRIVI IL TUO CODICE QUI
    massimo = numeri[0]
    for valore in numeri:
        if valore > massimo:
            massimo = valore
    return massimo
````

````python
# ✅ Test Esercizio 2.3 — dovresti vedere due righe con "✅ Corretto!"
casi = [([5, 3, 9, 1, 7], 9), ([-4, -1, -8], -1)]
for lista, atteso in casi:
    risultato = trova_massimo(lista)
    if risultato == atteso:
        print(f"✅ Corretto!  trova_massimo({lista}) → {risultato}")
    else:
        print(f"❌ trova_massimo({lista}) → {risultato}  (atteso: {atteso})")
````

## Parte 3 — Ciclo `while`

Il ciclo `while` si usa quando vuoi ripetere un'azione **finché una condizione rimane vera**, senza sapere in anticipo quante iterazioni saranno necessarie.

```python
while condizione:
    # blocco eseguito finché la condizione è True
    # ⚠️  ricorda sempre di modificare qualcosa
    #     altrimenti il ciclo non termina mai!
```

> **Attenzione:** assicurati sempre che la condizione del `while` possa diventare `False` ad un certo punto, altrimenti il programma si blocca in un ciclo infinito.

### Esercizio 3.1 — Conto alla rovescia

Scrivi una funzione `countdown(n)` che riceve un numero intero positivo e restituisce una lista con il conto alla rovescia da `n` fino a `1` (incluso).

Esempio: `countdown(5)` → `[5, 4, 3, 2, 1]`

**Suggerimenti:**
- Qui devi costruire una lista passo dopo passo.
- Scegli una condizione di ciclo che si fermi quando il conteggio ha raggiunto `1`.
- A ogni iterazione il valore corrente deve avvicinarsi alla condizione di stop.

````python
def countdown(n):
    # SCRIVI IL TUO CODICE QUI
    lista_numeri = []
    while n != 0:
        lista_numeri.append(n)
        n -= 1
    
    return lista_numeri
    
````

````python
# ✅ Test Esercizio 3.1 — dovresti vedere due righe con "✅ Corretto!"
casi = [(5, [5, 4, 3, 2, 1]), (3, [3, 2, 1])]
for n, atteso in casi:
    risultato = countdown(n)
    if risultato == atteso:
        print(f"✅ Corretto!  countdown({n}) → {risultato}")
    else:
        print(f"❌ countdown({n}) → {risultato}  (atteso: {atteso})")
````

### Esercizio 3.2 — Quanti passi per raggiungere la soglia?

Scrivi una funzione `passi_per_soglia(passo, soglia)` che simula un accumulo: partendo da `0`, aggiunge `passo` ad ogni iterazione e conta **quanti passi** sono necessari per raggiungere o superare `soglia`.

Esempio: `passi_per_soglia(10, 50)` → `5` (0→10→20→30→40→50: cinque incrementi)

**Suggerimenti:**
- Devi tenere traccia sia del totale raggiunto sia del numero di incrementi effettuati.
- Il ciclo deve continuare finché la soglia non è stata raggiunta.
- Aggiorna entrambe le informazioni a ogni iterazione.

````python
def passi_per_soglia(passo, soglia):
    # SCRIVI IL TUO CODICE QUI
    n = 0
    incrementi = 0
    while n != soglia:
        n += passo
        incrementi += 1
    return incrementi
    
        
    
````

````python
# ✅ Test Esercizio 3.2 — dovresti vedere due righe con "✅ Corretto!"
casi = [(10, 50, 5), (7, 21, 3)]
for passo, soglia, atteso in casi:
    risultato = passi_per_soglia(passo, soglia)
    if risultato == atteso:
        print(f"✅ Corretto!  passi_per_soglia({passo}, {soglia}) → {risultato}")
    else:
        print(f"❌ passi_per_soglia({passo}, {soglia}) → {risultato}  (atteso: {atteso})")
````

## Parte 4 — Funzioni

Le funzioni permettono di racchiudere un blocco di logica in un nome riutilizzabile. Si definiscono con `def`.

```python
def nome_funzione(parametro1, parametro2=valore_default):
    # corpo della funzione
    return risultato
```

- I **parametri** sono variabili nella definizione della funzione.
- Gli **argomenti** sono i valori concreti passati quando la chiami.
- `return` restituisce un valore al codice chiamante.
- Un parametro con `=` ha un **valore di default**: viene usato se non si passa un argomento esplicito.

### Esercizio 4.1 — Conversione di temperatura

Scrivi una funzione `celsius_fahrenheit(gradi_c)` che converte una temperatura da **Celsius** a **Fahrenheit** e restituisce il risultato.

La formula è:

$$F = C \times \frac{9}{5} + 32$$

**Suggerimenti:**
- Traduci la formula matematica in un'espressione Python usando il parametro della funzione.
- La funzione deve restituire il valore convertito.
- Non è necessario arrotondare il risultato.

````python
def celsius_fahrenheit(gradi_c):
    # SCRIVI IL TUO CODICE QUI
    gradi_c = (gradi_c * (9/5)) + 32
    return gradi_c
````

````python
# ✅ Test Esercizio 4.1 — dovresti vedere tre righe con "✅ Corretto!"
casi = [(0, 32.0), (100, 212.0), (37, 98.6)]
for c, atteso in casi:
    risultato = celsius_fahrenheit(c)
    if round(risultato, 1) == round(atteso, 1):
        print(f"✅ Corretto!  celsius_fahrenheit({c}) → {risultato}")
    else:
        print(f"❌ celsius_fahrenheit({c}) → {risultato}  (atteso: {atteso})")
````

### Esercizio 4.2 — Media di una lista

Scrivi una funzione `calcola_media(numeri)` che riceve una lista di numeri e restituisce la loro **media aritmetica**.

La formula è:

$$\text{media} = \frac{\text{somma degli elementi}}{\text{numero di elementi}}$$

**Suggerimenti:**
- La formula richiede due informazioni sulla lista: il totale e la quantità di elementi.
- Puoi usare le funzioni built-in viste finora quando non sono vietate dall'esercizio.
- Restituisci il risultato della divisione.

````python
def calcola_media(numeri):
    # SCRIVI IL TUO CODICE QUI
    totale = len(numeri)
    media = sum(numeri)/totale
    return media
````

````python
# ✅ Test Esercizio 4.2 — dovresti vedere due righe con "✅ Corretto!"
casi = [([2, 4, 6, 8], 5.0), ([10, 20, 30], 20.0)]
for lista, atteso in casi:
    risultato = calcola_media(lista)
    if risultato == atteso:
        print(f"✅ Corretto!  calcola_media({lista}) → {risultato}")
    else:
        print(f"❌ calcola_media({lista}) → {risultato}  (atteso: {atteso})")
````

### Esercizio 4.3 — Calcola il prezzo con IVA (parametro di default)

Scrivi una funzione `calcola_prezzo_iva(prezzo, iva=22)` che riceve un prezzo e una percentuale IVA, e restituisce il **prezzo totale** comprensivo di IVA.

La formula è:

$$\text{totale} = \text{prezzo} + \left(\text{prezzo} \times \frac{\text{iva}}{100}\right)$$

Il parametro `iva` ha un valore di **default pari a 22**, quindi se non viene specificato si usa l'IVA ordinaria italiana.

**Suggerimenti:**
- Scomponi il problema in due parti: quota IVA e prezzo finale.
- Usa il parametro di default per gestire il caso ordinario, ma lascia la funzione flessibile per altre percentuali.
- Restituisci il totale, non solo l'importo dell'IVA.

````python
def calcola_prezzo_iva(prezzo, iva=22):
    # SCRIVI IL TUO CODICE QUI
    totale = prezzo + (prezzo * (iva/100))
    return totale
````

````python
# ✅ Test Esercizio 4.3 — dovresti vedere tre righe con "✅ Corretto!"
# Il terzo test usa iva=4 (aliquota ridotta) al posto del default 22
casi = [(100, 22, 122.0), (200, 22, 244.0), (100, 4, 104.0)]
for prezzo, iva, atteso in casi:
    risultato = calcola_prezzo_iva(prezzo, iva)
    if risultato == atteso:
        print(f"✅ Corretto!  calcola_prezzo_iva({prezzo}, iva={iva}) → {risultato}")
    else:
        print(f"❌ calcola_prezzo_iva({prezzo}, iva={iva}) → {risultato}  (atteso: {atteso})")
````
