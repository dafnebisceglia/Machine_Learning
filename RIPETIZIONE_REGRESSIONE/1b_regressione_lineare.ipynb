## 5. Scrivere la funzione della retta

Ora scriviamo la funzione che calcola le previsioni del modello:

$$f_{w,b}(x) = wx + b$$

La funzione rappresenta una retta, ma nel codice la chiameremo `f_wb`, come nella notazione matematica.

### Task

Completa la funzione `f_wb`. Deve ricevere `x`, `w` e `b`, e restituire le previsioni.

```python
def f_wb(x, w, b):
    # Suggerimento: puoi moltiplicare direttamente un array NumPy per un numero
    return w * x + b
```

## 8. Fittare manualmente la retta

Ora prova a modificare `w` e `b` per far passare la retta il più vicino possibile ai punti.

### Range consigliati

Per non scegliere valori completamente fuori scala, resta inizialmente dentro questi intervalli:

- `w` tra `7000` e `12000`;
- `b` tra `15000` e `40000`.

Suggerimento pratico:

- se la retta cresce troppo lentamente, aumenta `w`;
- se la retta cresce troppo velocemente, diminuisci `w`;
- se la retta è tutta troppo in basso, aumenta `b`;
- se la retta è tutta troppo in alto, diminuisci `b`.

```python
# Modifica solo questi due valori e riesegui la cella.
w_scelto = 7000
b_scelto = 15000

previsioni_scelte = f_wb(x, w_scelto, b_scelto)

plt.scatter(x, y, color='red', marker='x', label='Dati reali')
plt.plot(x, previsioni_scelte, color='blue', label='Retta scelta')
plt.title('Tentativo manuale')
plt.xlabel('Anni di esperienza')
plt.ylabel('Salario annuo')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

print('w scelto:', w_scelto)
print('b scelto:', b_scelto)

# Domande guida:
# - La retta passa vicino alla maggior parte dei punti?
# - La retta sembra troppo inclinata o troppo piatta?
# - La retta sembra troppo alta o troppo bassa?
```

## 9. Confrontare più tentativi

Puoi provare più coppie `(w, b)` e disegnare più rette sullo stesso grafico. In questo modo puoi confrontarle visivamente e scegliere quella che segue meglio i dati.

```python
# Modifica solo queste coppie di valori e riesegui la cella.
tentativi = [
    (8000, 30000),
    (9000, 25000),
    (10000, 20000),
]

plt.scatter(x, y, color='red', marker='x', label='Dati reali')

for w_tentativo, b_tentativo in tentativi:
    previsioni_tentativo = f_wb(x, w_tentativo, b_tentativo)
    etichetta = f'w={w_tentativo}, b={b_tentativo}'
    plt.plot(x, previsioni_tentativo, label=etichetta)

plt.title('Confronto tra più rette')
plt.xlabel('Anni di esperienza')
plt.ylabel('Salario annuo')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

Modifica `anni_esperienza` e osserva come cambia la previsione.
