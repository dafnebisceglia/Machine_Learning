# Dataset 1 – Boston Housing

Il dataset **Boston Housing** contiene dati socio-economici, ambientali e strutturali raccolti in varie zone dell'area urbana di Boston negli anni '70. Le **features** ($X$) descrivono le caratteristiche dei quartieri — come il tasso di criminalità, la qualità delle scuole, i livelli di inquinamento e le infrastrutture abitative. **MEDV** è la variabile target ($y$), ovvero il valore economico medio delle abitazioni.

| Feature Name | Feature Description |
| --- | --- |
| **CRIM** | Tasso di criminalità pro capite per città. |
| **ZN** | % di terreni residenziali suddivisi in zone per lotti molto grandi (oltre 25.000 sq. ft.). |
| **INDUS** | % di acri commerciali non al dettaglio (zone industriali) per città. |
| **CHAS** | **One-Hot Encoded**: `1` se il distretto confina con il fiume Charles, altrimenti `0` |
| **NOX** | Concentrazione di ossidi di azoto nell'aria (misura l'inquinamento atmosferico). |
| **RM** | Numero medio di stanze per abitazione. |
| **AGE** | % di case abitate dai proprietari che sono state costruite prima del 1940. |
| **DIS** | Distanza ponderata rispetto ai cinque principali centri per l'impiego di Boston. |
| **RAD** | Indice di accessibilità alle grandi autostrade radiali (più alto = più vicino). |
| **TAX** | Aliquota dell'imposta immobiliare a valore pieno per ogni $10.000. |
| **PTRATIO** | Rapporto alunni-insegnanti nelle scuole del distretto. |
| **B** | Indice storico calcolato matematicamente sulla percentuale di residenti afroamericani. |
| **LSTAT** | % di popolazione appartenente a una fascia socio-economica più bassa. |
| **MEDV** | Valore mediano delle case occupate dai proprietari in migliaia di dollari (**Variabile Target**). |

# Dataset 2 – California Housing

Il dataset **California Housing** contiene dati socio-economici raccolti a livello di singoli isolati (block groups) in California durante il censimento del 1990. Le **features** ($X$) descrivono indicatori socio-economici, demografici e geografici della zona, mentre **MedHouseVal** è la variabile target ($y$), ovvero il valore medio delle abitazioni espresso in centinaia di migliaia di dollari.

| Feature Name | Feature Description |
| --- | --- |
| **MedInc** | Reddito medio dei residenti nel blocco censuario. |
| **HouseAge** | Età media delle abitazioni nel blocco. |
| **AveRooms** | Numero medio di stanze per abitazione. |
| **AveBedrms** | Numero medio di stanze da letto per abitazione. |
| **Population** | Popolazione totale residente nel blocco. |
| **AveOccup** | Numero medio di componenti per nucleo familiare (occupazione media). |
| **Latitude** | Latitudine geografica del blocco. |
| **Longitude** | Longitudine geografica del blocco. |
| **MedHouseVal** | **Variabile Target**: Valore medio delle case (in centinaia di migliaia di $). |

# Dataset 3 – Wine

Il dataset **Wine** contiene i risultati di un'analisi chimica su vini prodotti in una specifica regione dell'Italia, derivati da tre diversi tipi di cultivar (vitigni). Le **features** ($X$) descrivono le quantità di 13 diversi costituenti chimici presenti nel vino (come alcol, acido malico, ceneri, magnesio, ecc.). Le tre classi di vitigni, convenzionalmente mappate e identificate nel dataset come `0`, `1`, `2` corrispondono rispettivamente a tre  varietà storiche piemontesi: `Barolo`, `Grignolino` e `Barbera`.

| Feature Name | Feature Description |
| --- | --- |
| **alcohol** | Percentuale di volume alcolico presente nel vino. |
| **malic_acid** | Quantità di acido malico (influenza l'acidità e il gusto fruttato). |
| **ash** | Contenuto di ceneri (il residuo inorganico dopo l'evaporazione). |
| **alcalinity_of_ash** | Alcalinità delle ceneri (misura l'equilibrio acido-base del residuo). |
| **magnesium** | Quantità di magnesio presente (minerale essenziale per la fermentazione). |
| **total_phenols** | Quantità totale di fenoli (composti che influenzano colore e sapore). |
| **flavanoids** | Concentrazione di flavonoidi (antiossidanti legati a colore e astringenza). |
| **nonflavanoid_phenols** | Fenoli non flavonoidi (altri composti fenolici secondari). |
| **proanthocyanidins** | Proantocianidine (tannini complessi responsabili della struttura del vino). |
| **color_intensity** | Intensità del colore del vino (misurata spettrofotometricamente). |
| **hue** | Tonalità o sfumatura del colore del vino (es. tendente al giallo o al rosso). |
| **od280/od315_of_diluted_wines** | Misura di assorbanza proteica (indica la densità e la stabilità del vino). |
| **proline** | Amminoacido più abbondante nel vino (legato alla maturazione dell'uva). |
| **target** | **Variabile Target**: Classe di appartenenza del vino. |

# Dataset 4 – Pima Indians Diabetes

Il dataset **Pima Indians Diabetes** raccoglie informazioni cliniche e diagnostiche fornite dal *National Institute of Diabetes and Digestive and Kidney Diseases*. Tutti i pazienti registrati in questo archivio sono donne di almeno 21 anni di età appartenenti alla comunità indigena dei Pima (vicino a Phoenix, Arizona). Le **features** ($X$) descrivono misure mediche predittive e dati anamnestici, utilizzati per prevedere la **variabile target** ($y$), ovvero l'insorgenza o meno del diabete entro un orizzonte temporale di 5 anni.

| Feature Name | Feature Description |
| --- | --- |
| **Pregnancies** | Numero di gravidanze portate a termine dalla paziente. |
| **Glucose** | Concentrazione di glucosio plasmatico a 2 ore da un test di tolleranza al glucosio orale. |
| **BloodPressure** | Pressione arteriosa diastolica (espressa in mm Hg). |
| **SkinThickness** | Spessore della piega cutanea del tricipite (misura espressa in mm, usata per stimare il grasso corporeo). |
| **Insulin** | Concentrazione di insulina sierica a 2 ore dall'esame (espressa in mu U/ml). |
| **BMI** | Indice di Massa Corporea (Body Mass Index), calcolato come peso in kg diviso per il quadrato dell'altezza in metri. |
| **DiabetesPedigreeFunction** | Funzione di punteggio che esprime la predisposizione genetica al diabete in base alla storia familiare. |
| **Age** | Età della paziente (espressa in anni). |
| **Outcome** | **Variabile Target**: `1` se il paziente contrarrà il diabete entro 5 anni, altrimenti `0` |
